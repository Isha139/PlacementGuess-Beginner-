```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('placement.csv')
```

```python
df = df.iloc[:, 1:]
```

```python
plt.scatter(df['cgpa'], df['iq'], c=df['placement'])
```

```python
# Extract input and output column 
x = df.iloc[:, :2]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
#Train_Test_Split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x, y, test_size = 0.2) #0.2 means took aside the 20% for testing purpose
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
#Scaling
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train) #Tranform elements between 0 to 1
```

```python
x_train
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
x_test
```

```python
# Train the model
#By using logistic regression classifier

from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
# Model Training
clf.fit(x_train, y_train) #fit is used for training model
```

```python
#Evaluate model
y_pred = clf.predict(x_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred, y_test)
```
